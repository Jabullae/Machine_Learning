## k-최근접 이웃 알고리즘
- 지도학습 알고리즘 중 하나이다.
- 굉장히 직관적이고 간단한 알고리즘이다.
- 어떤 데이터가 주어지면 그 주변(이웃)의 데이터를 살펴본 뒤 많은 데이터가 포함되어 있는 범주로 분류하는 방식이다.
- knn의 특징은 훈련이 따로 필요 없다는 것이다. 다른 모델들은 fit(x_train, y_train)의 코드를 통해 훈련을 시킨다.
- 즉, 훈련데이터를 기반으로 모델을 만들고 테스트 데이터로 테스트하는 방식이지만 knn은 훈련데이터를 저장하는게 훈련의 전부이다.
- 모델을 별도로 구축하지 않는다는 뜻으로 게으른 모델(Lazy model)이라고 부른다.

### knn 알고리즘 : 2진 분류 알고리즘

```python
from sklearn.neighbors import KNeighborsClassifier
# 인스턴스 객체 생성
knn = KNeighborsClassifier()
# 훈련 - 모델에 데이터를 전달하여 규칙을 학습하는 과정을 말한다.
model = knn.fit(fish_data, fish_target)
# 평가(정확도 검사)
model.score(fish_data, fish_target)
```

- KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
-                     metric_params=None, n_jobs=None, n_neighbors=5, p=2,
-                     weights='uniform') <- 원래는 이렇게 결과가 나온다. 
- weights는 가중치. uniform은 균등하게, distance는 거리별로 가중치를 둔다는 의미
- n_neighbor는 k값. k값이 너무 적으면 과소적합, k값이 너무 크면 과대적합의 문제가 생긴다.
- knn은 적절한 k값을 찾는 것이 중요하다.
- p는 1일때 맨하튼거리, 2일때 유클리드 거리

### 피처 스케일링
- 두 특성의 스케일(scale)이 다를 때, 데이터를 표현하는 기준이 다르면 알고리즘이 올바르게 예측할 수 없다. 
- 최소 - 최대 정규화(min-max normalization) : X<sub>new</sub> = X - min(x) / max(x) - min(x)
- z-점수 표준화(z_value standardization) : X<sub>new</sub> = X-mean(x) / std(x)

```python
mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0)
```

```python
train_scaled = (train_input - mean) / std
```

```python
new = ([25, 150] - mean) / std
plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(new[0], new[1], marker = '^', color = 'g')
plt.title('훈련용 데이터(표준화)')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0)
test_scaled = (test_input - mean) / std
```

```python
kn.fit(train_scaled, train_target)
```

```python
kn.score(test_scaled, test_target)
```

```python
kn.predict([new])
```

```python
distances, indexs = kn.kneighbors([new])
plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(new[0], new[1], marker = '^', color = 'g')
plt.scatter(train_scaled[indexs, 0], train_scaled[indexs, 1], marker = 'D')
plt.title('훈련용 데이터(표준화)')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 최적의 k값 찾기

```python
from sklearn.metrics import accuracy_score

# 1 ~ 100

ks = 100
mean_acc = np.zeros((ks-1))

for k in range(1, ks):
    clf = KNeighborsClassifier(n_neighbors = k).fit(X_train, y_train)
    yhat = clf.predict(X_test)
    mean_acc[k-1] = accuracy_score(y_test, yhat)
    
print(mean_acc)
```

```python
# 시각화 
plt.plot(range(1, ks), mean_acc, 'g')
plt.xlabel('Number of Neighbors(k)')
plt.ylabel('Accuracy')
plt.tight_layout()
plt.show()
```

```python
clf = KNeighborsClassifier(n_neighbors = 20).fit(X_train, y_train)
clf.fit(X_train, y_train)
accuracy_score(y_test, clf.predict(X_test))
```

## 유방암 진단 데이터
- 유방암(breast cancer) 진단 데이터 세트
- 유방암 진단 사진으로부터 측정한 종양(tumar)의 특정값을 사용하여 종양이 양성(bengin)인지 악성(malignant)인지를 판별하는 예제

### 유방암 데이터 가져오기

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

### 훈련 데이터와 테스트 데이터로 분리

```python
print('cancer의 키 : \r', cancer.keys())
```

```python
# 데이터프레임으로 변환하여 내용 확인
df = pd.DataFrame(cancer.data, columns = cancer.feature_names)
sy = pd.Series(cancer.target, dtype = 'category')
sy = sy.cat.rename_categories(cancer.target_names)  # 0, 1를 각 종양의 양성, 음성으로 구분
df['cancer'] = sy
df.head(10)
```

```python
df.iloc[0, :]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cancer['data'], cancer['target'], random_state = 42)
```

```python
print('X_train 크기 : ', X_train.shape)
print('y_train 크기 : ', y_train.shape)
```

```python
print('X_test 크기 : ', X_test.shape)
print('y_test 크기 : ', y_test.shape)
```

### 데이터 살펴보기

### knn 최근접 이웃 알고리즘

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
print('훈련세트의 정확도 : {:.2f}'.format(knn.score(X_train, y_train)))
```

```python
from sklearn.metrics import accuracy_score

# 1 ~ 100

ks = 100
mean_acc = np.zeros((ks-1))

for k in range(1, ks):
    clf = KNeighborsClassifier(n_neighbors = k).fit(X_train, y_train)
    yhat = clf.predict(X_test)
    mean_acc[k-1] = accuracy_score(y_test, yhat)
    
print(mean_acc)
mean_acc.argmax()
```
