# 교차검증

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

## ggplot 스타일
mpl.style.use('ggplot')
# matplotlib 한글화
mpl.rcParams['font.family'] = 'D2coding'
```

```python
wine = pd.read_csv('https://raw.githubusercontent.com/rickiepark/hg-mldl/master/wine.csv')

wine_input = wine[['alcohol', 'sugar', 'pH']].to_numpy()
wine_target = wine['class'].to_numpy()
```

## 훈련세트와 테스트세트 분리
- X_train : 훈련 input
- X_test : 테스트 input
- y_train : 훈련 target
- y_test : 테스트 target

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine_input, wine_target, test_size = 0.2, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 훈련세트에서 검증세트를 분리
- 우리가 직접 분리하진 않지만 개념을 이해하기 위해 분리
- X_sub : 학습 input
- X_val : 학습 target
- y_sub : 검증 input
- y_val : 검증 target

```python
X_sub, X_val, y_sub, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)
```

## 모델 구축(DecisionTreeClassfier)

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 42) # seed값 부여
dt.fit(X_sub, y_sub)

print('학습 데이터 점수 : {}'.format(dt.score(X_sub, y_sub)))
print('검증 데이터 점수 : {}'.format(dt.score(X_val, y_val)))

# 과대적합이 되어있다는 것을 알수있다. 
```

## 교차 검증(k-fold cross validation)

```python
from sklearn.model_selection import cross_validate

# 평가할 모델 객체를 매개변수로 전달한다. 검정 세트를 떼어내지 않고 훈련 세트전체를 전달해야 한다.
scores = cross_validate(dt, X_train, y_train) # 폴드의 갯수를 늘리고 싶을때는 cv = ? 로 파라미터를 부여(default는 5개)
scores

# fit_time : 훈련시간, score_time : 검증시간, test_score : 교차검증점수
```

```python
print('교차검증한 점수의 평균 : {}'.format(np.mean(scores['test_score'])))
```

```python
from sklearn.model_selection import StratifiedKFold
scores = cross_validate(dt,  X_train, y_train, cv = StratifiedKFold())
np.mean(scores['test_score'])
```

```python
# 만약 k-fold가 10이라면(10-겹 교차검증)
splitter = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 42)
scores = cross_validate(dt,  X_train, y_train, cv = splitter)
np.mean(scores['test_score'])
```

## 그리드 서치

```python
from sklearn.model_selection import GridSearchCV
model = DecisionTreeClassifier(random_state = 42)

params = {
    'max_depth':[2, 3, 4, 5, 6, 7, 8, 9, 10],   
    'min_samples_leaf':[1, 2, 3, 4, 5, 6, 7, 8, 9],
    'min_samples_split':[2, 3, 4, 5, 6, 7, 8, 9, 10]
}

gs = GridSearchCV(model, param_grid = params, cv = 5, refit = True, n_jobs = -1) # n_jobs는 실행 코어수 조절(-1을 주면 모든 코어 사용)
gs.fit(X_train, y_train)
```

```python
# 그리드 서치를 사용한 모델의 최적의 파라미터값
model = gs.best_estimator_
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
# 그리드 서치를 통해 찾은 최적의 매개변수
print(gs.best_estimator_)
print(gs.best_params_)
print(gs.best_score_)
```

```python
# 최상의 교차검증 점수
np.max(gs.cv_results_['mean_test_score'])
```

```python
# 교차검증을 안함
model = DecisionTreeClassifier(max_depth = 8, min_samples_leaf = 8, min_samples_split = 2, random_state = 42)
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize = (30, 20))
plot_tree(model, filled = True, feature_names = ['alcohol', 'sugar', 'pH'], class_names=['Red', 'White'])
plt.show()
```

```python
from sklearn.tree import export_graphviz
import graphviz

export_graphviz(model, out_file = 'tree2.dot', class_names = ['red', 'wine'], 
               feature_names = ['alcohol', 'sugar', 'pH'], impurity = True, filled = True)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
with open('tree2.dot') as f:
    dot_graph = f.read()
    
src = graphviz.Source(dot_graph)
src
```
