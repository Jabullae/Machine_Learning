```python
# 각 feature에 대한 분포를 시각화하는 작업
def plot_hist(df):
    plt.rcParams['figure.figsize'] = [20, 19]
    fig = plt.figure(1) # 도화지 한장 이라는 의미
    
    for i in range(len(df.columns)):
        ax = fig.add_subplot(5, 5, i+1)
        plt.hist(df[df.columns[i]], bins = 50)
        ax.set_title(df.columns[i])
        
    plt.show()
```

## 전처리 : 피처 스케일링

- 피처 스케일링 : 피처들간의 단위를 맞추는 작업
- StandardScaler : 표준화(평균 = 0, 표준편차 = 1)

```python
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x : (x-series_mean) / series_std)
        
    return df
```

```python
# 회귀분석 : 독립변수와 종속변수를 분리하는 작업
# 종속변수 : 연봉(2018)
# 독립변수를 사용할 feature들을 리스트로 저장
scale_columns = ['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR',
       '연봉(2017)']

picher_df = standard_scaling(picher, scale_columns)
```

```python
picher_df = picher_df.rename(columns = {'연봉(2018)':'y'})
picher_df.head()
```

### 피처 스케일링 : 더미 or 원-핫 인코딩

- 문자 또는 범주형 데이터를 정형화된 데이터(0과 1)로 변환하는 작업

```python
team_encoding = pd.get_dummies(picher_df['팀명'])
# 기존에 팀명 컬럼 삭제
```

```python
picher_df = picher_df.drop('팀명', axis = 1)
picher_df = picher_df.join(team_encoding)
```

```python
picher_df.head()
```

```python
# 성능 평가 함수
from sklearn.metrics import mean_squared_error, mean_absolute_error

# MAE, RMSE
def evaluate_rger(y, pred):
    # MAE
    mae = mean_absolute_error(y, pred)
    
    # RMSE
    rmse = np.sqrt(mean_squared_error(y, pred))

    print(f'MAE : {mae:.3f}', f'RMSE : {rmse:.3f}')
```

## 선형회귀

### 기본 예측/평가

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = picher_df[picher_df.columns.difference(['선수명', 'y'])]
y = picher_df['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)

# 평가함수 호출
evaluate_rger(y_test, pred)
```

## 예측값, 실제값 데이터 분석

```python
def get_top_error_data(y_test, pred, n_tops = 5):
    # 예측값, 실제값 데이터 프레임 생성
    result_df = pd.DataFrame(y_test.values, columns = ['real_y'])
    result_df['pred_y'] = np.round(pred)
    result_df['diff'] = np.abs(result_df['real_y'] - result_df['pred_y'])
    
    # 예측값과 실제값의 차이가 큰 데이터 순서대로 추출
    print(result_df.sort_values('diff', ascending=False)[:n_tops])
    
get_top_error_data(y_test, pred, n_tops = 10)
```

```python
import statsmodels.api as sm

# 상수항
X_train = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train).fit()
model.summary()
```

## 피처들의 영향력을 파악하기 위한 시각화 - summary의 params

```python
# 회귀계수를 리스트로 변환하는 작업
coefs = model.params.to_list()
coefs_series = pd.Series(coefs)

# 변수명(index)를 리스트로 변환하는 작업
x_labels = model.params.index.to_list()

plt.figure(figsize = (7, 10))
ax = coefs_series.plot(kind = 'barh')
ax.set_title('컬럼들의 상관계수 : 영향력있는 컬럼 추출')
ax.set_ylabel('features')
ax.set_xlabel('coef')
ax.set_yticklabels(x_labels)
plt.show()
```

## 피처들의 상관관계를 분석하는 시각화 - heatmap

```python
# 상관계수
corr = picher_df[scale_columns].corr(method='pearson')
plt.figure(figsize = (15, 15))
sns.heatmap(corr, annot = True, cmap = 'Blues', linewidth = 0.5)
plt.show()
```

## 회귀분석의 성능을 향상시키기 위해 다중공선성 확인

```python
# 분산 팽차 요인(VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 피처들의 vif 계수 확인
vif = pd.DataFrame()
vif['VIF factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['features'] = X.columns
vif.round(1)
```

## 적절한 피처를 사용하여 다시 모델 학습

```python
X = picher_df[['FIP', 'WAR', '볼넷/9', '삼진/9', '선발', '연봉(2017)']]
y = picher_df['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# LinearRegression
lr = LinearRegression()
model = lr.fit(X_train, y_train)
pred = lr.predict(X_test)

# 평가함수 호출
evaluate_rger(y_test, pred)
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```
