# 로지스틱 회귀로 와인 분류하기 - 화이트 와인

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

## ggplot 스타일
mpl.style.use('ggplot')
# matplotlib 한글화
mpl.rcParams['font.family'] = 'D2coding'
```

- https://raw.githubusercontent.com/rickiepark/hg-mldl/master/wine.csv

```python
wine = pd.read_csv('https://raw.githubusercontent.com/rickiepark/hg-mldl/master/wine.csv')
wine.head()
```

```python
wine.shape
```

```python
wine.info()
```

```python
wine['class'].unique()
```

```python
wine.describe()
```

```python
wine['class'].value_counts()
```

```python
wine_input = wine[['alcohol', 'sugar', 'pH']].to_numpy()
wine_target = wine['class'].to_numpy()
```

```python
wine_input.shape, wine_target.shape
```

# 훈련 세트와 테스트 세트로 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine_input, wine_target, test_size = 0.2, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 피쳐 스케일링

```python
# 표준화를 이용한 스케일링
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X_train) # X_train에서 표준화할 데이터를 찾으면 test_input은 fit을 할 필요가 없다.
train_scaled = ss.transform(X_train)
test_scaled = ss.transform(X_test)
```

## 로지스틱회귀

```python
# 로지스틱 회귀
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, y_train)
print(lr.score(train_scaled, y_train))
print(lr.score(test_scaled, y_test))
```

```python
lr.classes_
```

```python
lr.predict(test_scaled[:20])
```

```python
lr.predict_proba(test_scaled[:20])
#    0(red wine),   1(white wine)
```

```python
wine.keys()
```

```python
print('회귀계수 : {}'.format(lr.coef_), 
      '\n절편 : {}'.format(lr.intercept_),
      '\n회귀방정식 : Z = {:.3f} + {:.3f} x alcohol + {:.3f} x sugar + {:.3f} x pH'.format(
          lr.intercept_[0], lr.coef_[0, 0], lr.coef_[0, 1], lr.coef_[0, 2]))
```

```python
# z값 계산
decisions = lr.decision_function(test_scaled[:30])
decisions
```

```python
# 시그모이드 함수를 통해 확률값 추출
from scipy.special import expit

# 양성일 확률(1일 확률)
print('양성일 확률(1일 확률) : {}'.format(expit(decisions)))

# 음성일 확률(0일 확률)
print('\n음성일 확률(0일 확률) : {}'.format(1 - expit(decisions)))
```

```python
# 시그모이드 함수 시각화
def sigmoid(x):
    return 1 / (1+np.exp(-x))

x_data = np.arange(-10, 10, 0.1)
result = sigmoid(x_data)
plt.figure(figsize=(12, 8))
plt.plot(x_data, result)
plt.scatter(decisions, expit(decisions), marker='^' ,color = 'g')
plt.scatter(0, 0.5, marker = 'D', color = 'black')
plt.axhline(0.5, 0, 1, color='black', linestyle='--', linewidth=2)
plt.title('시그모이드 시각화', size = 20)
plt.xlabel('X_data', size = 20)
plt.ylabel('result', size = 20)
plt.xlim(-8, 8)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 의사결정나무

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 42) # seed값 부여
dt.fit(train_scaled, y_train)

print('훈련용 데이터 점수 : {}'.format(dt.score(train_scaled, y_train)))
print('테스트 데이터 점수 : {}'.format(dt.score(test_scaled, y_test)))

# 훈련용 데이터 점수가 상대적으로 높기때문에 과대적합을 의심해 볼 수 있다. 
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (25, 50))
plot_tree(dt)
plt.show()
```

```python
# 의사결정나무는 균일트리를 만드는 것을 목적
plt.figure(figsize = (30, 20))
plot_tree(dt, max_depth = 2, filled = True, feature_names = ['alcohol', 'sugar', 'pH'])
plt.show()
```

- 지니 불순도 : 1-{(음성클래스의 비율)<sup>2</sup> + (양성클래스의 비율)<sup>2</sup>}
- 불순도가 작을수록 정보 획득량은 증가한다.

```python
# root 노드의 지니 불순도 구하기
1 - ((1258/5197)**2 + (3939/5197)**2)
```

## 가지치기(pruning)
- 최대 트리로 결정된 결정트리의 특정 노드 밑의 트리를 제거하여 일반화 성능을 높이는 것
- 과적합을 막기위해 사용된다.
- 더 많은 가지가 생기지 않도록 최대 깊이, leaf 노드의 최대개수, 한 노드가 분할하기 위한 최소 데이터 수를 제한한다.
- 가지가 많으면 깊이가 깊어지고 깊이가 깊어지면 과대적합이 발생한다.
- 성능이 뛰어나지 않다. 그래서 단독적으로 쓰이지 않고 딥러닝이 나온 이후에는 더욱 사용하지 않는다.

```python
dt = DecisionTreeClassifier(max_depth = 3, random_state = 42) # seed값 부여
dt.fit(train_scaled, y_train)

print('훈련용 데이터 점수 : {}'.format(dt.score(train_scaled, y_train)))
print('테스트 데이터 점수 : {}'.format(dt.score(test_scaled, y_test)))
```

```python
# 의사결정나무는 균일트리를 만드는 것을 목적
plt.figure(figsize = (30, 20))
plot_tree(dt, filled = True, feature_names = ['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
dt = DecisionTreeClassifier(max_depth = 3, random_state = 42) # seed값 부여
dt.fit(X_train, y_train)

print('훈련용 데이터 점수 : {}'.format(dt.score(X_train, y_train)))
print('테스트 데이터 점수 : {}'.format(dt.score(X_test, y_test)))
```

```python
plt.figure(figsize = (30, 20))
plot_tree(dt, filled = True, feature_names = ['alcohol', 'sugar', 'pH'], class_names=['Red', 'White'])
plt.show()
```

```python
# 특성 중요도 : 결정트리는 어떤 특성이 가장 유용한지를 나타내는 특성 중요도를 계산해 준다.
print('특성 중요도 - alcohol : {:.3f}, sugar : {:.3f}, pH : {:.3f}'.format(
    dt.feature_importances_[0], dt.feature_importances_[1], dt.feature_importances_[2]))
```
